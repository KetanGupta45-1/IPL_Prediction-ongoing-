```python
df["Bowling Style"] = df["Bowler"].map(bowling_styles)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder_bowler = LabelEncoder()
encoder_venue = LabelEncoder()
encoder_striker = LabelEncoder()
encoder_innings = LabelEncoder()
encoder_style = LabelEncoder()
```

```python
df['Bowler_ID'] = encoder_bowler.fit_transform(df['Bowler'])
df['Venue_ID'] = encoder_venue.fit_transform(df['Venue'])
df['Striker_ID'] = encoder_striker.fit_transform(df['Striker'])
df['Innings_ID'] = encoder_innings.fit_transform(df['Innings'])
df['Bowl_style_ID'] = encoder_style.fit_transform(df['Bowling Style'])
```

```python
df.head()
```

```python
df['Over_Number'] = df['Over'].astype(int)
df['Ball_Number'] = ((df['Over'] - df['Over_Number']) * 10).astype(int)

window_size = 5  # Last 5 balls average
df['Rolling_Avg_Runs'] = df.groupby('Striker_ID')['Runs by Batsman'].transform(lambda x: x.rolling(window=window_size, min_periods=1).mean())

df['Batsman_Bowler_Combo'] = df['Striker_ID'].astype(str) + '_' + df['Bowler_ID'].astype(str)
df['Batsman_Bowler_Combo'] = LabelEncoder().fit_transform(df['Batsman_Bowler_Combo'])
```

```python
strike_rate_df = df.groupby(['Striker_ID', 'Bowler_ID']).agg({'Runs by Batsman': 'sum', 'Over': 'count'}).reset_index()
strike_rate_df['Strike_Rate'] = (strike_rate_df['Runs by Batsman'] / strike_rate_df['Over'])

df = df.merge(strike_rate_df[['Striker_ID', 'Bowler_ID', 'Strike_Rate']], on=['Striker_ID', 'Bowler_ID'])
```
